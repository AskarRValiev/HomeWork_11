```python
from sympy import *
x = Symbol('x')
y = Symbol('y')
f = Symbol('f')
f = (sin(x)/2 + cos(x)/2)**2 * sin(x)
y = solve((sin(x)/2 + cos(x)/2)**2 * sin(x))
y.append(7*pi/4)
y = sorted(y)
print(y)
```

```python
Найти вершины: Необходимое условие существования экстремума функции. 
Если в некоторой точке x функция f имеет экстремум, то в этой точке производная равна нулю.
```

```python
extr = [] #создаем список экстремумов
for i in range (-45, 316):
    x = i*pi/180
    p = (-sin(x) + cos(x))*(sin(x)/2 + cos(x)/2)*sin(x) + (sin(x)/2 + cos(x)/2)**2*cos(x)
    ext = p.evalf()
    if ext < 0.008 and ext > -0.008:
        extr.append(x.evalf(n = 3))
        print(f'Экстремум в точке {x.evalf(n = 3)} радиан или {i} градусов')
print(extr)
 
    
        
```

```python
Найти интервалы, на которых функция возрастает и убывает
```

```python
for i in range(len(extr)-1):
    x = (extr[i] + extr[i+1])/2
    if (-sin(x) + cos(x))*(sin(x)/2 + cos(x)/2)*sin(x) + (sin(x)/2 + cos(x)/2)**2*cos(x) < 0:
        print(f'функция убывает на промежутке от {extr[i]} радиан до {extr[i+1]} радиан')
    else:
        print(f'функция возрастает на промежутке от {extr[i]} радиан до {extr[i+1]} радиан')
            
```
